```python
import string
string.punctuation
```

```python
exclude  = string.punctuation
def remove_punctuation(text):
    for char in exclude:
        text = text.replace(char, '')
    return text 
```

```python
text = "string. with. punctuation?"
print(remove_punctuation(text))
```

```python
# Another method to remove the punctuation from the string

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

remove_punctuation("string. with. punctuation?")
```

```python
df['review'] = df['review'].apply(remove_punctuation)
```

```python
import nltk
# Corrected stopword removal function
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    return " ".join([word for word in text.split() if word not in stop_words])

# Apply stopword removal to the reviews
df['review'] = df['review'].apply(remove_stopwords)
```

```python
#Handling emojis
import re

def remove_emoji(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"
                               # emoticons
                               u"\U0001F300-\U0001F5FF"
                               # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"
                               # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"
                               # flags (105)
                               u"\U00002702-\U00002780"
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
# Example usage
text_with_emoji = "Hello! 😀 This is a sample text with emoji. 🌟. Today i am happy 😊❤"
text_without_emoji = remove_emoji(text_with_emoji)
print(text_without_emoji) 
```

```python
# Tokenization and Stemming
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem_words(text):
    return " ".join([ps.stem(word) for word in word_tokenize(text)])

# Example of stemming a sentence
stemmed_sentence = stem_words('I loved loving it')
print(stemmed_sentence)

# Apply stemming to the reviews
df['review'] = df['review'].apply(stem_words)
```

```python
# Handling NaN or empty reviews before vectorization
df = df[df['review'].notnull()]
df = df[df['review'] != '']
```

```python
X=df.iloc[:,0:1].values
X.shape
```

```python
# CountVectorizer to convert text data into numerical format
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=1000, stop_words='english')  # Change stop_words to 'english'
X = cv.fit_transform(df['review']).toarray()

# Output the shape of the transformed data
print(X.shape)
```

```python
X[0].mean()
```

```python
y=df.iloc[:,-1].values
y.shape
y = df['sentiment'].values  # Change to ensure you're using the sentiment column
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
X_train.shape
X_test.shape
y_train.shape
y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_train_scl = scaler.fit_transform(X_train)
X_test_scl = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model_dt = DecisionTreeClassifier(random_state=42)
model_dt.fit(X_train_scl, y_train)
```

```python
#Accuracy of the model on training and testing data
 
print("Training Accuracy :", model_dt.score(X_train_scl, y_train))
print("Testing Accuracy :", model_dt.score(X_test_scl, y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(random_state=42)
model_rf.fit(X_train_scl, y_train)
```

```python
#Accuracy of the model on training and testing data
 
print("Training Accuracy :", model_rf.score(X_train_scl, y_train))
print("Testing Accuracy :", model_rf.score(X_test_scl, y_test))
```

```python
import xgboost
from xgboost import XGBClassifier
model_xgb = XGBClassifier(random_state=42)
model_xgb.fit(X_train_scl, y_train)
```

```python
#Accuracy of the model on training and testing data
 
print("Training Accuracy :", model_xgb.score(X_train_scl, y_train))
print("Testing Accuracy :", model_xgb.score(X_test_scl, y_test))
```
